## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
## Write your code here
# Identify the number of duplicate rows
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")

# Display the duplicate rows
duplicate_rows = df[df.duplicated()]
print(duplicate_rows)
```

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
## Write your code here
# 1. Identify all rows that are part of a duplicate group based on the subset
subset_cols = ['MainBranch', 'Employment', 'RemoteWork']
# keep=False ensures all members of the duplicate groups are included
duplicates_mask = df.duplicated(subset=subset_cols, keep=False)
df_duplicates = df[duplicates_mask]

print(f"Number of rows in duplicate groups based on {subset_cols}: {len(df_duplicates)}")

# 2. Analyze consistency in other columns
# We calculate the number of unique values per group for each column.
# If a column has 1 unique value in a group, it is identical within that duplicate group.
analysis = df_duplicates.groupby(subset_cols).nunique()

# 3. Calculate the frequency of 'identicalness'
# We find the percentage of groups where each column has only 1 unique value
identical_freq = (analysis == 1).mean() * 100

# Sort to see which columns are most frequently identical
identical_freq_sorted = identical_freq.sort_values(ascending=False)

print("\nPercentage of duplicate groups where the column contains identical values:")
print(identical_freq_sorted.head(10))
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
## Write your code here
# --- Visualization 1: Top 10 Countries with Duplicates (Bar Chart) ---
plt.figure(figsize=(10, 6))
country_counts = df_duplicates['Country'].value_counts().head(10)

sns.barplot(x=country_counts.values, y=country_counts.index, palette='viridis')
plt.title('Top 10 Countries by Number of Duplicate Rows')
plt.xlabel('Number of Duplicates')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig('duplicates_by_country.png')
plt.show()

# --- Visualization 2: Distribution by Employment (Pie Chart) ---
plt.figure(figsize=(8, 8))
emp_counts = df_duplicates['Employment'].value_counts()

plt.pie(emp_counts, labels=emp_counts.index, autopct='%1.1f%%', 
        startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Duplicates by Employment Status')
plt.tight_layout()
plt.savefig('duplicates_by_employment.png')
plt.show()

print(f"Total duplicates found: {len(df_duplicates)}")
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
## Write your code here
# 2. Define your critical columns (The "Fingerprint")
critical_columns = ['MainBranch', 'Country', 'Age', 'EdLevel', 'YearsCode']

# 3. Identify and remove duplicates based on the subset
# keep='first' ensures we keep the first occurrence and delete subsequent ones
df_cleaned = df.drop_duplicates(subset=critical_columns, keep='first')

# 4. Verification
original_count = len(df)
cleaned_count = len(df_cleaned)
removed = original_count - cleaned_count

print(f"Original Row Count: {original_count}")
print(f"Cleaned Row Count: {cleaned_count}")
print(f"Logical Duplicates Removed: {removed}")
```
