**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Demo: Running an SQL Query


Count the number of rows in the table named 'main'


```python
QUERY = """
SELECT COUNT(*) 
FROM main
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: Listing All Tables


To view the names of all tables in the database:


```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
# Query data
QUERY = "SELECT CompTotal FROM main"
df_comp = pd.read_sql_query(QUERY, conn)

# Plot Histogram
plt.figure(figsize=(10, 5))
sns.histplot(df_comp['CompTotal'].dropna(), bins=50, kde=True)
plt.title('Distribution of Total Compensation')
plt.xlabel('Total Compensation')
plt.ylabel('Frequency')
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
# Query data
QUERY = "SELECT Age, ConvertedCompYearly FROM main"
df_age_comp = pd.read_sql_query(QUERY, conn)

# Plot Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Age', y='ConvertedCompYearly', data=df_age_comp)
plt.title('Box Plot of Annual Compensation by Age Group')
plt.xticks(rotation=45)
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
# Query data
QUERY = "SELECT WorkExp, ConvertedCompYearly FROM main"
df_scatter = pd.read_sql_query(QUERY, conn)

# Plot Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='WorkExp', y='ConvertedCompYearly', data=df_scatter)
plt.title('Relationship between Work Experience and Annual Compensation')
plt.xlabel('Years of Work Experience')
plt.ylabel('Annual Compensation (USD)')
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here
# Here we use the count of respondents as the bubble size.
QUERY = """
SELECT WorkExp, AVG(ConvertedCompYearly) as AvgComp, COUNT(*) as RespondentCount
FROM main
GROUP BY WorkExp
"""
df_bubble = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(12, 8))
sns.scatterplot(x='WorkExp', y='AvgComp', size='RespondentCount', sizes=(20, 2000), data=df_bubble, alpha=0.5)
plt.title('Bubble Plot: Work Experience vs Avg Compensation (Size = Number of Respondents)')
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
# Query and process data (Splitting multi-select strings)
QUERY = "SELECT LanguageHaveWorkedWith FROM main"
df_lang = pd.read_sql_query(QUERY, conn)
languages = df_lang['LanguageHaveWorkedWith'].str.split(';', expand=True).stack().value_counts()

# Plot Pie Chart for Top 5
plt.figure(figsize=(8, 8))
languages.head(5).plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Languages Worked With')
plt.ylabel('') # Remove y-label
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
## Write your code here
QUERY = "SELECT Age, RemoteWork, COUNT(*) as Count FROM main GROUP BY Age, RemoteWork"
df_stacked = pd.read_sql_query(QUERY, conn)
df_pivot = df_stacked.pivot(index='Age', columns='RemoteWork', values='Count')

df_pivot.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Remote Work Composition by Age Group')
plt.ylabel('Number of Respondents')
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
QUERY = "SELECT EdLevel, ConvertedCompYearly FROM main"
df_bar = pd.read_sql_query(QUERY, conn)
median_comp = df_bar.groupby('EdLevel')['ConvertedCompYearly'].median().sort_values()

plt.figure(figsize=(10, 6))
median_comp.plot(kind='barh')
plt.title('Median Annual Compensation by Education Level')
plt.xlabel('Median Compensation (USD)')
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here
QUERY = "SELECT EdLevel, ConvertedCompYearly FROM main"
df_bar = pd.read_sql_query(QUERY, conn)
median_comp = df_bar.groupby('EdLevel')['ConvertedCompYearly'].median().sort_values()

plt.figure(figsize=(10, 6))
median_comp.plot(kind='barh')
plt.title('Median Annual Compensation by Education Level')
plt.xlabel('Median Compensation (USD)')
plt.show()
```
