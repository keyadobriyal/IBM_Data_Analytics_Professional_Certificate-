### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
##Write your code here
sat_age_composition = df.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Creating the stacked bar chart
sat_age_composition.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median Job Satisfaction (6 & 7) Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Median Satisfaction Points')
plt.legend(title='Satisfaction Metric')
plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
##Write your code here
sat_emp_composition = df.groupby('Employment')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Creating the stacked bar chart
sat_emp_composition.plot(kind='bar', stacked=True, figsize=(10, 6), color=['skyblue', 'orange'])
plt.title('Median Job Satisfaction by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Median Points')
plt.xticks(rotation=45)
plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
##Write your code here
age_filter = df[df['Age'].isin(['25-34 years old', '35-44 years old'])]

# Aggregating median values for comparison
comp_sat_3035 = age_filter.groupby('Age')[['ConvertedCompYearly', 'JobSatPoints_6']].median()

# Plotting the stacked chart (Note: Scaling differs significantly between USD and points)
comp_sat_3035.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median Compensation and Satisfaction (Ages ~30-35)')
plt.show()
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python
##Write your code here
comp_sat_age = df.groupby('Age')[['ConvertedCompYearly', 'JobSatPoints_6']].median()

# Creating the stacked bar chart
comp_sat_age.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Median Compensation and Job Satisfaction Across All Age Groups')
plt.ylabel('Values (Stacked)')
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
##Write your code here
db_age_df = df[['Age', 'DatabaseWantToWorkWith']].dropna()
db_age_df['Database'] = db_age_df['DatabaseWantToWorkWith'].str.split(';')
db_age_df = db_age_df.explode('Database')

# Taking the Top 5 most preferred databases for a cleaner visualization
top_dbs = db_age_df['Database'].value_counts().nlargest(5).index
db_age_top = db_age_df[db_age_df['Database'].isin(top_dbs)]

# Pivoting for stacked representation
db_pivot = db_age_top.groupby(['Age', 'Database']).size().unstack().fillna(0)
db_pivot.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Top 5 Preferred Databases by Age Group')
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
##Write your code here
emp_sat_dist = df.groupby(['JobSatPoints_6', 'Employment']).size().unstack().fillna(0)

# Plotting the stacked chart
emp_sat_dist.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Employment Type Distribution per Job Satisfaction Level')
plt.xlabel('Job Satisfaction Score (0-10)')
plt.ylabel('Number of Respondents')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
##Write your code here
# Exploding multi-valued LanguageAdmired column
lang_age_df = df[['Age', 'LanguageAdmired']].dropna()
lang_age_df['Language'] = lang_age_df['LanguageAdmired'].str.split(';')
lang_age_df = lang_age_df.explode('Language')

# Focusing on Top 5 languages
top_langs = lang_age_df['Language'].value_counts().nlargest(5).index
lang_pivot = lang_age_df[lang_age_df['Language'].isin(top_langs)].groupby(['Age', 'Language']).size().unstack().fillna(0)

lang_pivot.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Top 5 Admired Programming Languages by Age Group')
plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
##Write your code here
plat_emp_df = df[['Employment', 'PlatformAdmired']].dropna()
plat_emp_df['Platform'] = plat_emp_df['PlatformAdmired'].str.split(';')
plat_emp_df = plat_emp_df.explode('Platform')

# Top 5 Platforms
top_plats = plat_emp_df['Platform'].value_counts().nlargest(5).index
plat_pivot = plat_emp_df[plat_emp_df['Platform'].isin(top_plats)].groupby(['Employment', 'Platform']).size().unstack().fillna(0)

plat_pivot.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Top 5 Admired Platforms by Employment Type')
plt.xticks(rotation=45)
plt.show()
```
